## Sec. 2.1 Analytic forms
**Theorem 2** (Marsaglia (2006); Pham-Gia et al. (2007)).  
For $W=X_1 / X_2$, there are constants $r$ and $s$ such that:  
$r(W-s)$ is distributed as $T=\frac{a+V_1}{b+V_2}$,  
$W$ is distributed as $\frac{1}{r} T+s$, where $s=\frac{\sigma_1}{\sigma_2} \rho, 1/r=\frac{\sigma_1}{\sigma_2} \sqrt{1-\rho^2}$  

$$
\begin{gathered}
f_W(w)=\frac{1}{\frac{\sigma_1}{\sigma_2} \sqrt{1-\rho^2}} f_T\left(\frac{w-\rho \frac{\sigma_1}{\sigma_2}}{\frac{\sigma_1}{\sigma_2} \sqrt{1-\rho^2}}\right), \\
\left(V_1, V_2\right) \sim \mathcal{B N}(0,0 ; 1,1,0), b=\frac{\mu_2}{\sigma_2} \geq 0, a= \pm \frac{\mu_1 / \sigma_1-\rho \mu_2 / \sigma_2}{\sqrt{1-\rho^2}} \geq 0, \\
f_T(t)=\frac{1}{\exp \left(\frac{a^2+b^2}{2}\right) \pi\left(1+t^2\right)}{ }_1 F_1\left(\begin{array}{c}
1 \\
1 / 2
\end{array}; \frac{(a t+b)^2}{2\left(1+t^2\right)}\right),
\end{gathered}
$$

where ${ }_1 F_1\left(\begin{array}{l}\alpha \\ \beta\end{array} ; z\right)$ is Kummer's confluent hypergeometric function.

**Refs.**
- Hanč J., Hančová M., Borovský D., Probability distributions and calculations for Hake’s ratio statistics in measuring effect size, 
*preprint submitted to Statistical Papers*
- Marsaglia, George. 2006. “Ratios of Normal Variables.” Journal of Statistical Software 16 (4). https://doi.org/10.18637/jss.v016.i04.
- Pham-Gia, T., N. Turkkan, and E. Marchand. 2007. “Density of the Ratio of Two Normal Random Variables and Applications.” Communications in Statistics - Theory and Methods 35 (9): 1569–91. https://doi.org/10.1080/03610920600683689.

```python
# variables
a, b, t, q = var('a,b,t,q', domain='real')

# Given that q = (b + a*t) / sqrt(1 + t^2) is the upper boundary from 0 to q,
# the following condition must also hold
assume(b + a*t > 0)
```

```python
# closed forms for hypergeometric functions
from sage.functions.hypergeometric import closed_form

# Individual terms for f_T according to Thm
C = exp(-(a^2+b^2)/2)
q = (b + a*t)/sqrt(1+t^2)
M(a,b,z) = hypergeometric_M(a,b,z).generalized()
f_q(t) = M(1,1/2, q^2/2)/(pi*(1+t^2))

# definition of f_T
f_T(t) = C*f_q(t)
show(f_T(t))
```

## Sec. 2.2 Shapes and approximations
### Modality

Regions of points $(a, b)$ for which the density of $T$ is unimodal or bimodal.
Unimodal if left of curve, bimodal if right.

```python
graphTitle = f'The modality curve for (a,b)'
bm(a) = (18.621 - 63.411*a + 84.041*a**2 - 
        54.668*a**3 + 17.716*a**4 - 2.2986*a**5) / (2.256058904 - a)

gm = plot([], figsize=(4,5), title=graphTitle, frame=True, gridlines=True, ymax=15)
gm += plot(bm(t), (t,1,2.256058904), color='blue')

# Add a vertical line at x=2.256058904 in red color
gm += line([(2.256058904, 0), (2.256058904, 15.5)], color='red')
gm += line([(1, 0), (1, 15.5)], color='red')

gm.show(ymin=0, xmin=0.5, xmax=2.5, ymax=15)
```

```python
# check the unimodal condition
is_unimodal= lambda values: bool(b.subs(values) > bm(a).subs(values))
```

Specifically, we considered $(a,b) = (2,0.25), (1.5,1), (4,7), (5,25)$, which were not chosen randomly. They represent four qualitatively distinct shape categories with respect to normality: 
- (1) $(2,0.25)$: *bimodal*, 
- (2) $(1.5,1)$: *unimodal with heavy tails* (bad normal approximation), 
- (3) $(4, 7)$: *unimodal, skewed* (reasonable normal approximation)
- (4) $(5,25) $: *unimodal, symmetric* (good normal approximation).	

### Plots of PDF for specific values

```python
# Layout of the plots
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')

# Parameters for the graph
graphTitle = ""
gf = None

def set_plot(values, color = 'blue', text = f'Density function of the ratio:'):
    # Declare that we are using the global variables
    global graphTitle, gf 
    # Construct the graph title
    graphTitle = text+f'${str(values).replace("==", "=")}$'
    # Create the plot (assuming `plot` is a predefined function in your context)
    gf = plot([], color=color, title=graphTitle, figsize=(8, 5), frame=True)
```

```python
# specific values 
a_val, b_val = 2, 1/4

# PDF of ratio
values = [a == a_val, b == b_val]
m, s, k = a/b, (a/b)*sqrt(1/a^2+1/b^2), 0.5
tmin, tmax = (m-k*s).subs(values),(m+k*s).subs(values)
fT(t) = f_T(t).subs(values)
fT.show()

# plot of fT
set_plot(values)
gf+= plot(lambda t:fT(t), (t, tmin, tmax))
gf
```

```python
# analytical unimodal condition
is_unimodal(values)
```

```python
# specific values 
a_val, b_val = 3/2, 1

# PDF of ratio
values = [a == a_val, b == b_val]
m, s, k = a/b, (a/b)*sqrt(1/a^2+1/b^2), 2
tmin, tmax = (m-k*s).subs(values),(m+k*s).subs(values)
fT(t) = f_T(t).subs(values)
fT.show()

# plot of fT
set_plot(values)
gf+= plot(lambda t:fT(t), (t, tmin, tmax))
gf
```

```python
# analytical unimodal condition
show(values)
is_unimodal(values)
```

According to Marsaglia (2006, sec. 3.1) the PDF could be theoretically bimodal, but in the practice is unimodal since the second mode may be remotely located and very small as to be insignificant.

```python
# positions of infinitesimal peak
dfTdt = diff(fT(t),t).simplify_hypergeometric().canonicalize_radical()
Cv = C.subs(values)
```

```python
Cv.n()
```

```python
tleft = (dfTdt/Cv).find_root(-100,0)
peak = (tleft, fT(tleft).n())
peak
```

```python
gf += points(peak,color='red',pointsize=50);
gf
```
