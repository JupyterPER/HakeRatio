# Hake ratio as an example

For ratio of independent normal variables:  $T = X_1/X_2$ 

- Numerator $ X_1 \sim N(a, 1) $:
- Denominator $ X_2 \sim N(b, 1) $:

## INTEGRAND terms for integration

### CF of $X_1$
$\varphi_{X_1}(t_1)$

**Hake ratio** - For ratio of indepedent normal variables in $T = X_1/X_2$

Numerator $ X_1 \sim N(a, 1), a = 1.5 $:
- CF: $ \varphi_{X_1}(t) = e^{i a t - \frac{t^2}{2}}$

Bivariate ratio CF of $X_1$
- CF$_d$: $ \dfrac{\varphi_{X_1}(t_1)}{t_2}$ 

```matlab
% characteristic function of X1 ~ N(a,1)
a = 1.5;
cf1 = @(t1) exp(1i * a * t1 - t1.^2 / 2);
```

```matlab
% bivariate cf1 divided by t2
cf1_d = @(t) cf1(t(:,1)) ./ t(:,2);
```

### Derivative of CF of $X_2$
$ \varphi_{X_2}'(t_2)$

**Hake ratio** - For ratio of indepedent normal variables in $T = X_1/X_2$

Denominator $ X_2 \sim N(b, 1) $, $ b = 1 $\:
- CF: $ \varphi_{X_2}(t) = e^{i b t - \frac{t^2}{2}}$
- CF derivative: $\varphi_{X_2}'(t) = \left( i b - t \right) e^{i  bt - \frac{t^2}{2}}$

```matlab
% derivative of characteristic function of X2 and its symbolic derivative
b = 1;
cf2 = @(t2) exp(1i * b* t2 - t2.^2 / 2);
cf2_dif = @(t2) (1i*b - t2) .* exp(1i * b* t2 - t2.^2 / 2);
```

```matlab
% Numerical derivative of X2 using finite differences
h = 1e-5;
cf2_dif_num = @(x) (cf2(x + h) - cf2(x - h)) / (2*h);
cf2_dif_num2 = @(x) (4/3 * (cf2(x + h) - cf2(x - h)) / (2*h)) - (1/3 * (cf2(x + 2*h) - cf2(x - 2*h)) / (4*h));
```

**Note: Accurate numerical derivative**  
*Midpoint approximation*
$$
q^{\prime}(x) = \frac{q(x+h) - q(x-h)}{2h} + O(h^2).
$$
*Richardson extrapolation*
$$
q^{\prime}(x) = \frac{4}{3} \frac{q(x+h) - q(x-h)}{2 h} - \frac{1}{3} \frac{q(x+2 h) - q(x-2 h)}{4 h} + O\left(h^4\right).
$$

# 2D Integration

**Section 3.2:** Inversion formula for ratio $T$ of independent RVs $X_1,X_2$
$$
f_T(x) \equiv f(x_1, x_2 = 1) = \frac{1}{\pi^2} \int_{-\infty}^{+\infty} \int_0^{+\infty} \Re\left\{\varphi_{X_2}^{\prime}(- x_1 t_1 - x_2 t_2)\,\frac{\varphi_{X_1}(t_1)}{t_2} \right\} d t_2 d t_1.
$$

Numerical quadrature (according to Mijanović et al 2023)
$$
f\left(x_1, x_2=1\right)\approx\frac{1}{\pi^2} h_1 h_2 \sum_{v_1=-N}^{N} \sum_{v_2=0}^{N}  \Re\left\{ \varphi_{X_2}^{\prime}(- x_1 b_1 - x_2 b_2) \frac{\varphi_{X_1}(b_1)}{b_2}\right\}$$

$$b_1=h_1\left(v_1+0.5\right), b_2=h_2\left(v_2+0.5\right)$$

**Refs.**
- Mijanović, Andjela, Božidar V. Popović, and Viktor Witkovský. 2023. “A Numerical Inversion of the Bivariate Characteristic Function.” Applied Mathematics and Computation 443:127807. https://doi.org/10.1016/j.amc.2022.127807.

- Original Matlab code for bivariate cfs  
https://github.com/witkovsky/CharFunTool/blob/master/CF_InvAlgorithms/cf2Dist2D.m



## INPUT PARAMETERS for integration

```matlab
% parameters for boundaries
SixSigmaRule = 6;
tolDiff = 1e-6;
% accuracy for cf functions
cftTol = 1e-14;
% tuning constant
k = 3;
```

## BOUNDARIES for integral

The mean and standard deviation of the random vector $X=\left(X_1, X_2\right)$ from the given marginal CFs using the higher-order methods for approximating the derivatives. In particular, we use the following estimates as Mijanović et al, 2023:

$$
\begin{aligned}
& \operatorname{mean}\left(X_1\right)=\left(\frac{8 \Im(\mathrm{cf} 1(\delta))}{5 \delta}-\frac{2 \Im(\mathrm{cf} 1(2 \delta))}{5 \delta}+\frac{8 \Im(\mathrm{cf} 1(3 \delta))}{105 \delta}-\frac{2 \Im(\mathrm{cf} 1(4 \delta))}{280 \delta}\right), \\
& \operatorname{mo}_2\left(X_1\right)=\left(\frac{205}{72 \delta^2}-\frac{16 \Re(\mathrm{cf} 1(\delta))}{5 \delta^2}+\frac{2 \Re(\mathrm{cf} 1(2 \delta))}{5 \delta^2}-\frac{16 \Re(\mathrm{cf} 1(3 \delta))}{315 \delta^2}+\frac{2 \Re(\mathrm{cf} 1(4 \delta))}{560 \delta^2}\right), \\
& \operatorname{std}\left(X_1\right)=\sqrt{\operatorname{mo}_2\left(X_1\right)-\text { mean }^2\left(X_1\right)}
\end{aligned}
$$

where $\delta$ is chosen tolerance for numerical differentiation (as a default value we use $\delta=10^{-4}$ ) and $\mathrm{mo}_2$ denotes the estimate of the second moment. Similarly, calculate mean $\left(X_2\right)$ and $\operatorname{std}\left(X_1\right)$ so that then mean $(X)=$ $\left[\operatorname{mean}\left(X_1\right)\right.$, mean $\left.\left(X_2\right)\right]$ and $\operatorname{std}(X)=\left[\operatorname{std}\left(X_1\right), \operatorname{std}\left(X_2\right)\right]$.

```matlab
% auxiliary functions for estimates of moments
cft1               = cf1(tolDiff*(1:4)');
cftRe1             = real(cft1);
cftIm1             = imag(cft1);
cft2               = cf2(tolDiff*(1:4)');
cftRe2             = real(cft2);
cftIm2             = imag(cft2);
```

```matlab
% first moments 
xMean(1) = (8*cftIm1(1)/5 - 2*cftIm1(2)/5 + 8*cftIm1(3)/105 ...
            - 2*cftIm1(4)/280) / tolDiff;
xMean(2) = (8*cftIm2(1)/5 - 2*cftIm2(2)/5 + 8*cftIm2(3)/105 ...
            - 2*cftIm2(4)/280) / tolDiff;
% second moments
xM2(1)   = (205/72 - 16*cftRe1(1)/5 + 2*cftRe1(2)/5 ...
    - 16*cftRe1(3)/315 + 2*cftRe1(4)/560) / tolDiff^2;
xStd(1)  = sqrt(xM2(1) - xMean(1)^2);
xM2(2)   = (205/72 - 16*cftRe2(1)/5 + 2*cftRe2(2)/5 ...
    - 16*cftRe2(3)/315 + 2*cftRe2(4)/560) / tolDiff^2;
xStd(2)  = sqrt(xM2(2) - xMean(2)^2);
```

```matlab
% boundaries and range
xMin = xMean - SixSigmaRule * xStd;
xMax = xMean + SixSigmaRule * xStd;
range = xMax - xMin;

xMin, xMax, range
```

```matlab
% set of values x for pdf
xN = 101;
x1 = linspace(xMin(1),xMax(1),xN);
```

```matlab
% parameters for a Riemann sum
N                  = 2^8; % Set large N to improve the precision, e.g. N = 2^10
dt                 = (2*pi/k) ./ range;
t1                 = (0.5+(-N:N))'*dt(1);
t2                 = (0.5+(0:N))'*dt(2);
```

```matlab
% Set integral limits for t1 and t2 for numerical integration
t1_min = min(t1);  % Approximate -infinity
t1_max = max(t1);  % Approximate +infinity
t2_min = 0;        % t2 starts from zero
t2_max = max(t2);  % Approximate +infinity
```

# Built-in 1D integrators

Inversion formula for ratio $T$ of independent RVs $X_1,X_2$ can be rewritten as 
$$
f_T(x) = \frac{1}{\pi^2}  \int_{-\infty}^{+\infty}  \Re\left\{ \varphi_{X_1}(t_1)\int_0^{+\infty} 
\frac{\varphi_{X_2}'(-t_2 - x t_1)}{t_2} d t_2 \right\} d t_1
$$

PDF as the iterated integration via the **B**uilt-in 1D integrator (`integral`)

```matlab
% accuracy settings
reg_tol = 1e-15;
rtol = 1e-5;

% Loop over each x value in the array
% Initialize an array to store fTx values for each x
fTx_values = zeros(size(x1));

parfor i = 1:length(x1)
    x = x1(i); % Current x value
    
    % Define f_t1, which integrates over t2 for each specific t1
    first_integral = @(t1) integral(@(t2) cf2_dif(-t2 - x * t1) ./ t2, reg_tol, t2_max, ...
                          'ArrayValued', true, 'RelTol', rtol);
    
    % Define the integrand for fTx
    integrand = @(t1) cf1(t1) .* first_integral(t1);
    
    % Calculate the outer integral for fTx
    fTx_values(i) = (1/pi^2) * real(integral(integrand, t1_min, t1_max, 'RelTol', rtol));
end

% Display the result
Rpdf_0_1D = max(0, fTx_values);
```

## Analytic form of PDF 
Pham-Gia 2007
- Pham-Gia, T., Turkkan, N., & Marchand, E. (2007). Density of the Ratio of Two Normal Random Variables and Applications. Communications in Statistics - Theory and Methods, 35(9), 1569–1591. https://doi.org/10.1080/03610920600683689

$$
f_T(t) = \frac{\exp\left(-\frac{a^2 + b^2}{2}\right)}{\pi (1 + t^2)}{}_1F_1\left(\begin{array}{c}
1 \\
1 / 2
\end{array}; q^2\right), \quad q = \frac{b + a t}{\sqrt{1 + t^2}}
$$


```matlab
% Define the function f_T for specific values
a = 1.5; b = 1;
f_k = @(t) exp(-(a^2 + b^2)/2) / (pi * (1 + t.^2));
q = @(t) (b + a * t) / sqrt(1 + t.^2);

% Define the function f_T for specific values
fT = @(t) f_k(t) .* Hypergeom1F1(1,1/2, q(t).^2 / 2);
```

```matlab
% Values of the function f_T for plot
t_values = linspace(-10, 10, 1000); % Define range for t
fT_values = arrayfun(fT, t_values);
```

### Comparing plots: analytic vs. 1D integrator

```matlab
% RPDF1 vs Analytic
figure
plot(t_values, fT_values, 'r', 'LineWidth', 2);
hold on;
plot(x1,Rpdf_0_1D,'b', 'LineWidth',1)
xlim([xMin(1) xMax(1)]) % plot in  sigma range
grid on,
title('Analytic PDF vs. Num Inv PDF - built-in 1D integrator')
xlabel('x')
ylabel('pdf')   
legend('f_T(t)', 'Rpdf_0_1_D');
```

# Built-in  2D integrators

Inversion formula for ratio $T$ of independent RVs $X_1,X_2$
$$
f_T(x) = \frac{1}{\pi^2} \int_0^{+\infty} \int_{-\infty}^{+\infty} \Re\left\{\frac{\varphi_{X_1}(t_1)}{t_2} \varphi_{X_2}^{\prime}(-t_2 - x t_1)\right\} d t_1 d t_2.
$$

PDF as a direct calculation via 2D integrators.

## Standard loop 

```matlab
% Define the integrand function
integrand = @(t1, t2, x) (cf1(t1) ./ t2) .* cf2_dif(-t2 - x .* t1);
```

```matlab
if Runloops
    if isempty(gcp('nocreate'))
        parpool;
    end
    % Suppress all warnings
    warning('off', 'all');

    % Perform the double integral for each x1
    % Initialize an array to hold the PDF values
    pdf = zeros(size(x1)); % Preallocate PDF array
    
    % Loop through each value of x = x1
    for i = 1:length(x1)
        xi = x1(i);
        fT_xi = quad2d(@(t1, t2) integrand(t1, t2, xi), ...
                      t1_min, t1_max, ...
                      t2_min, t2_max, ...
                      'RelTol', 1e-3, ...
                      'AbsTol', 1e-5  ...
                       );
        pdf(i) = (1 / pi^2) * real(fT_xi);
    end
    % Ensure ratio PDF values are non-negative
    Rpdf_0_quad = max(0, pdf);
    % Re-enable warnings if needed
    warning('on', 'all');
end
```

```matlab
if Runloops
    if isempty(gcp('nocreate'))
        parpool;
    end
    % Suppress all warnings
    warning('off', 'all');
    % Perform the double integral for each x1
    % Initialize an array to hold the PDF values
    pdf = zeros(size(x1)); % Preallocate PDF array
    
    %options = optimset('AbsTol', 1e-4, 'RelTol', 1e-3); % Adjust tolerances
    
    % Loop through each value of x = x1
    parfor i = 1:length(x1)
        xi = x1(i);
        fT_xi = integral2(@(t1, t2) integrand(t1, t2, xi), ...
                      t1_min, t1_max, ...
                      t2_min, t2_max, ...
                      'RelTol', 1e-3, ...
                      'AbsTol', 1e-5  ...
                       );
        pdf(i) = (1 / pi^2) * real(fT_xi);
    end
    % Ensure ratio PDF values are non-negative
    Rpdf_0_integr = max(0, pdf);
    % Re-enable warnings if needed
    warning('on', 'all');
end
```

## Loop using Parallelization

```matlab
if Runloops
    if isempty(gcp('nocreate'))
        parpool;
    end
    % Suppress all warnings
    warning('off', 'all');

    % Perform the double integral for each x1
    % Initialize an array to hold the PDF values
    pdf = zeros(size(x1)); % Preallocate PDF array
    
    % Loop through each value of x = x1
    parfor i = 1:length(x1)
        xi = x1(i);
        fT_xi = quad2d(@(t1, t2) integrand(t1, t2, xi), ...
                      t1_min, t1_max, ...
                      t2_min, t2_max, ...
                      'RelTol', 1e-3, ...
                      'AbsTol', 1e-5  ...
                       );
        pdf(i) = (1 / pi^2) * real(fT_xi);
    end
    % Ensure ratio PDF values are non-negative
    Rpdf_0_integr_par = max(0, pdf);
    % Re-enable warnings if needed
    warning('on', 'all');
end
```
